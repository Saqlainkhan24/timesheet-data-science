```python
import pandas as pd
import os

# List of Excel file paths
file_paths = [
    r'c:\Users\Admin\Desktop\excel\0102024-rptdailytsemp (3).xlsx', r'c:\Users\Admin\Desktop\excel\0202024-rptdailytsemp.xlsx',
    r'c:\Users\Admin\Desktop\excel\0302024-rptdailytsemp.xlsx', r'c:\Users\Admin\Desktop\excel\0402024-rptdailytsemp.xlsx',
    r'c:\Users\Admin\Desktop\excel\0502024-rptdailytsemp.xlsx', r'c:\Users\Admin\Desktop\excel\0602024-rptdailytsemp.xlsx' ,
    r'c:\Users\Admin\Desktop\excel\0702024-rptdailytsemp.xlsx', r'c:\Users\Admin\Desktop\excel\0802024-rptdailytsemp.xlsx'
]

output_dir = 'modified_files/'

for file_path in file_paths:

    df = pd.read_excel(file_path, header=None)

    df = df.iloc[1:-2]

    df.columns = df.iloc[0]

    df = df[1:]
     
    output_file_path = os.path.join(output_dir, f"{os.path.basename(file_path)}")
        
    df.to_excel(output_file_path, index=False, header=True)
        
    print(f"Processed {file_path} -> {output_file_path}")

    print(df)

print("Processing complete.")
```

```python
import pandas as pd
import calendar
from datetime import datetime

project_mapping = pd.read_excel(r'first_file_with_modified_planned_loading3.xlsx')  # Update with actual path or DataFrame

df = pd.read_excel(r'concatenated3_output.xlsx')

#Convert the Hours column from string format to timedelta
df['Hours'] = pd.to_timedelta(df['Hours'] + ':00')

df['date'] = pd.to_datetime(df['Date']).dt.strftime('%m-%y')

# Group by Name and Year-Month and sum the Hours (as timedelta)
monthly_timesheet = df.groupby(['Employee Name','date','Project Name'])['Hours'].sum().reset_index()

# Convert the timedelta back to a string format (hours:minutes)
monthly_timesheet['Hours'] = monthly_timesheet['Hours'].apply(lambda x: f"{int(x.total_seconds() // 3600)}:{int((x.total_seconds() % 3600) // 60):02d}")

# Drop the timedelta column if not needed
#monthly_timesheet = monthly_timesheet.drop(columns=['Hours'])

# Merge the timesheet with the project mapping dataset on the 'Project Name' column
monthly_timesheet = pd.merge(monthly_timesheet, project_mapping, how='outer')

#monthly_timesheet.to_excel('outputs_file9.xlsx', index=False)

print(monthly_timesheet)
```
